```python
x=np.arange(1,11)
y=3*x+6
plt.title('Matplotlib Demo')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x,y)
plt.show()
```

```python
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.title('Sine wave')
plt.show()
```

```python
#bar plot

x=[2,8,10]
y=[11,16,9]
plt.bar(x,y,align='center')
plt.title('bar graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

# pie chart

```python
labels='Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['Gold','Yellowgreen','lightcoral','lightskyblue']
explode=(0.3,0,0.2,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
